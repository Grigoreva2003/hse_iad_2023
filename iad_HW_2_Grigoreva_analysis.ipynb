#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
info_name = """Data/Students_info_{}.csv"""
marks_name = """Data/Students_marks_{}.csv"""

df = pd.DataFrame()
for i in range(10):
    info_i = pd.read_csv(info_name.format(i))
    marks_i = pd.read_csv(marks_name.format(i))

    class_i = pd.merge(info_i, marks_i, on="index")
    df = pd.concat([df, class_i])
display(df.sample(10))
display(df.shape)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
%%time

race_dict = {}
for row in df.values:
    race_dict.setdefault(row[1], []).append(row[-2])

for key in race_dict.keys():
    race_dict[key] = sum(race_dict[key]) / len(race_dict[key])

race_dict = sorted(race_dict.items(), key=lambda x: x[0])
pd.DataFrame(race_dict)
```

```python
%%time

ed_dict = {}
for row in df.values:
    if row[2] not in ed_dict.keys():
        ed_dict[row[2]] = row[-1]
    else:
        ed_dict[row[2]] = min(row[-1], ed_dict[row[2]])

ed_dict = sorted(ed_dict.items(), key=lambda x: x[0])
pd.DataFrame(ed_dict)
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def score(points):
    return "P" if points >= passmark else "F"

df["Math_PassStatus"] = df["math score"].apply(score)
df["Reading_PassStatus"] = df["reading score"].apply(score)
df["Writing_PassStatus"] = df["writing score"].apply(score)
df.head()
```

```python
types = ["math score", "reading score", "writing score"]
statuses = ["Math_PassStatus", "Reading_PassStatus", "Writing_PassStatus"]

for i in range(3):
    passed = df[statuses[i]].value_counts()
    print("Number of students who passed " + types[i].split()[0], '=', passed["P"])
    print("Number of students who didn't pass " + types[i].split()[0], '=', passed["F"])
    print()
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def overall(row):
    return "P" if row[-1] == "P" and row[-2] == "P" and row[-3] == "P" else "F"

df["OverAll_PassStatus"] = df.apply(overall, axis=1)
df.head()
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
        return "A"
    if 80 <= average_mark <= 90:
        return "B"
    if 70 <= average_mark <= 80:
        return "C"
    if 60 <= average_mark <= 70:
        return "D"
    if 50 <= average_mark <= 60:
        return "E"
    return "F"

df["Grade"] = ((df["math score"] + df["reading score"] + df["writing score"]) / 3).apply(GetGrade)
df
```
